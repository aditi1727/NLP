## Import Library

## Open JSON File

```python
import json

# Loading json file
f = open('g_news.json')
data = json.load(f)

# Obtaining title and description dictonaries from json file
title_dict = data['title']
desc_dict = data['desc']

title_text = []
desc_text = []
# Retriving the text of title and description dictonaries and storing it into variables in string format
for key,value in title_dict.items():
    title_text.append(value)
    
for key,value in desc_dict.items():
    desc_text.append(value)
    
title_text = str(title_text)
desc_text = str(desc_text)

text = title_text + desc_text
text
```

## Lowering the case

```python
text = text.lower()
text
```

## Remove the Punctuation marks in the text

```python
import string
print(string.punctuation)
```

```python
text = "".join([char for char in text if char not in string.punctuation+"’"+"‘"])
text
```

## Tokenizing the sentences

```python
from nltk import word_tokenize

words = word_tokenize(text)
words
```

## Removing the StopWord from the text

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

filtered_words = [word for word in words if word not in stop_words]
filtered_words
```

## Filtering the StopWords

```python
from nltk.stem.porter import PorterStemmer

porter = PorterStemmer()
stemmed = [porter.stem(word) for word in filtered_words]
```

```python
from nltk import pos_tag

pos = pos_tag(filtered_words)
pos
```

```python
len(filtered_words)
```

## Stemming the StopWords

```python
from nltk.stem import WordNetLemmatizer
  
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
lemmatized_words
```

## N - Gram Model

```python
from nltk import bigrams,trigrams
from collections import Counter,defaultdict
```

## Unigram of the model

```python
uni = {w1: 0 for i in lemmatized_words for w1 in i}
```

```python
#Counting frequency
for sentence in lemmatized_words:
    for w1 in sentence:
        uni[w1] += 1
print(uni)
```

```python
total_count = float(sum(uni.values()))
for w1 in uni:
    uni[w1] /= total_count
uni
```

```python
model_uni = {k: v for k,v in sorted(uni.items(),key=lambda item :item[1],reverse=True)}
model_uni
```

## Bigram of the Model

```python
big = list(bigrams(lemmatized_words))
big
```

```python
model_bigr = defaultdict(lambda: defaultdict(lambda:0))
```

```python
#Counting frequency
for sentence in lemmatized_words:
    for w1,w2 in bigrams(sentence,pad_right=True, pad_left=True):
        model_bigr[(w1)][w2] += 1
```

```python
#Estimating probability
for w1 in model_bigr:
    total_count = float(sum(model_bigr[w1].values()))
    for w2 in model_bigr[w1]:
        model_bigr[w1][w2] /= total_count
```

```python
dict(model_bigr["tea"])
```

## Trigram of the model

```python
tri = list(trigrams(lemmatized_words))
tri
```

```python
model_trigr = defaultdict(lambda: defaultdict(lambda:0))
```

```python
#Counting frequency
for sentence in lemmatized_words:
    for w1,w2,w3 in trigrams(sentence,pad_right=True, pad_left=True):
        model_trigr[(w1,w2)][w2] += 1
```

```python

#Estimating probability
for w1_w2 in model_trigr:
    total_count = float(sum(model_trigr[w1_w2].values()))
    for w3 in model_trigr[w1_w2]:
        model_trigr[w1_w2][w3] /= total_count
```

```python
dict(model_trigr["white","tea"])
```

## Finding Random Probability

```python
from copy import deepcopy as dc
```

```python
def findprob(m1,m2,text):
    text=text.split()
    text.insert(0,None)
    prob=1
    m1[None]=1
    for i in range(1,len(text)):
        if text[i] not in m1:
            temp=(m2[text[i-1]][text[i]]+1)/(len(m1))
            prob*=temp
        else:
            temp=(m2[text[i-1]][text[i]]+1)/(m1[text[i]]+len(m1))
            prob*=temp
    return prob
```

```python
text="black tea"
findprob(dc(model_uni),dc(model_bigr),text)
```
